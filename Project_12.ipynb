```python
#this is going to be the normalization for the sleep levels
def sleep_levels(row): #Defining the function with the name energy_levels
    if row['Sleep_efficiency'] <=.50: 
        #we are going to et a static value to make it simple, 
        #for low we have to use a value that falls between 0-50
        return .45
    elif row['Sleep_efficiency'] > .51 and row['Sleep_efficiency'] <.70: 
        return  .65
    else: #We assume all other values are in the middle, so we assign medium to them
        return .90
```

```python

df['sleep_levels'] = df.apply(sleep_levels,axis=1)

# Check to see how your function did
df[['Sleep_efficiency','sleep_levels']].head()
```
